# 15.You are offered a position either as ‘Sales Representative’ or ‘Marketing Representative’. You want to join a position with a smaller number of employees. Which position will you join?

```python
SR = list(jobs.loc[jobs.job_title =='Sales Representative'][['job_id']].job_id)[0]
MR = list(jobs.loc[jobs.job_title =='Marketing Representative'][['job_id']].job_id)[0]
```

```python
job = employee.groupby(employee.job_id).size().reset_index()
```

```python
job[(job.job_id.isin([SR,MR]))]
```
